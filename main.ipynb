```python
import dotenv
import os

from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from pydantic import SecretStr
from langchain_chroma import Chroma
from src.runnables.query_variation import QueryVariation
from src.runnables.retrieval import MessageRetrieval
from src.runnables.answer import FormulateAnswer
```

```python
dotenv.load_dotenv()
```

```python
embeddings = OpenAIEmbeddings(
    base_url=os.getenv("BASE_EMBEDDING_URL"),
    api_key=SecretStr("abc"),
    model=os.getenv("EMBEDDING_MODEL"),
    tiktoken_enabled=False,
)

vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory="./chroma_db",
)

llm = ChatOpenAI(
    base_url=os.getenv("BASE_LLM_URL"),
	api_key=SecretStr("abc"),
    model=os.getenv("LLM_MODEL"),
    temperature=0,
)
```

```python
query_variation = QueryVariation(llm_instance=llm)
retrieval = MessageRetrieval(vector_store=vector_store, embeddings=embeddings)
formulate = FormulateAnswer(llm_instance=llm)
```

```python
chain = query_variation | retrieval | formulate
```
